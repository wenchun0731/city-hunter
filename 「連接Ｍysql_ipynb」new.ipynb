```python
import pandas as pd
from datetime import datetime

# 獲取當前時間並格式化
current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

text = 'BAA1234'  # 車牌
data = {
    'license_plate': [text],
    'date_time': [current_time],  # 使用當下時間
    'location': ['TAIPEI'],
    'violation': ['闖紅燈'],  # 填闖紅燈或未禮讓行人
    'seconds': [5.00],  # 違規當下的秒數
}

df = pd.DataFrame(data)
```

```python
# SQL插入语句
add_violation = ("INSERT INTO testapp_car "
                 "(license_plate, date_time, location, violation, seconds) "
                 "VALUES (%s, %s, %s, %s, %s)")

# 将DataFrame中的每一行数据插入到数据库中
for index, row in df.iterrows():
    data_violation = (row['license_plate'], row['date_time'], row['location'], row['violation'], row['seconds'])
    cursor.execute(add_violation, data_violation)

# 提交事务
cnx.commit()
```

```python
# 关闭游标和连接
cursor.close()
cnx.close()
```
