```python
#測試判斷
!pip install ultralytics
!pip install deep-sort-realtime
import numpy as np
import cv2
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort
from collections import deque
from google.colab.patches import cv2_imshow

model = YOLO('yolov8l.pt')

# 打開視頻文件
cap = cv2.VideoCapture('/content/drive/MyDrive/0807.mov')  # 輸入影片
fps = cap.get(cv2.CAP_PROP_FPS)
size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
fNUMS = cap.get(cv2.CAP_PROP_FRAME_COUNT)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
videoWriter = cv2.VideoWriter("/content/drive/MyDrive/test-video03.mp4", fourcc, fps, size) # 輸出影片

# 初始化DeepSort跟蹤器
tracker = DeepSort(max_age=30)

# 緩存幀的隊列，用於保存前5秒的視頻
frame_buffer = deque(maxlen=int(fps * 5))  # 假設前5秒的緩存

def box_label(image, box, label='', color=(128, 128, 128), txt_color=(0, 255, 0)):
    p1, p2 = (int(box[0]), int(box[1])), (int(box[2]), int(box[3]))
    cv2.rectangle(image, p1, p2, color, thickness=1, lineType=cv2.LINE_AA)
    if label:
        w, h = cv2.getTextSize(label, 0, fontScale=2 / 3, thickness=1)[0]
        outside = p1[1] - h >= 3
        p2 = p1[0] + w, p1[1] - h - 3 if outside else p1[1] + h + 3
        cv2.rectangle(image, p1, p2, color, -1, cv2.LINE_AA)
        cv2.putText(image,
                    label, (p1[0], p1[1] - 2 if outside else p1[1] + h + 2),
                    0, 2/3, txt_color, thickness=1, lineType=cv2.LINE_AA)
        # 在框框上方顯示座標
        coord_label = f"({int(box[0])}, {int(box[1])}) - ({int(box[2])}, {int(box[3])})"
        cv2.putText(image, coord_label, (p1[0], p1[1] - 22 if outside else p1[1] + h + 22),
                    0, 2/3, txt_color, thickness=1, lineType=cv2.LINE_AA)

def save_video(buffer, output_path, fps, size, fourcc):
    out = cv2.VideoWriter(output_path, fourcc, fps, size)
    while buffer:
        out.write(buffer.popleft())
    out.release()

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # 添加當前幀到緩存
    frame_buffer.append(frame.copy())

    results = model(frame, conf=0.4)
    outputs = results[0].boxes.data.cpu().numpy()

    detections = []
    red_light_detected = False
    stop_line_detected = False
    car_cross_stop_line = False

    if outputs is not None:
        for output in outputs:
            x1, y1, x2, y2 = list(map(int, output[:4]))
            label = int(output[5])
            # 假設類別2是車，類別5是公交車，類別7是卡車，類別10是紅綠燈，類別11是停止線
            if label == 2:
                detections.append(([x1, y1, int(x2-x1), int(y2-y1)], output[4], 'car'))
            elif label == 5:
                detections.append(([x1, y1, int(x2-x1), int(y2-y1)], output[4], 'bus'))
            elif label == 7:
                detections.append(([x1, y1, int(x2-x1), int(y2-y1)], output[4], 'truck'))
            elif label == 10:  # 紅綠燈
                red_light_detected = output[6] == 0  # 假設紅燈標記為0
            elif label == 11:  # 停止線
                stop_line_detected = True
                stop_line = (x1, y1, x2, y2)

        tracks = tracker.update_tracks(detections, frame=frame)

        for track in tracks:
            if not track.is_confirmed():
                continue
            track_id = track.track_id
            bbox = track.to_ltrb()

            # 檢查車輛是否超過停止線
            if stop_line_detected and red_light_detected:
                car_center_x = (bbox[0] + bbox[2]) // 2
                car_center_y = (bbox[1] + bbox[3]) // 2
                if car_center_y > stop_line[1]:  # 檢查車輛是否超過停止線
                    car_cross_stop_line = True

            box_label(frame, bbox, '#' + str(int(track_id)) + ' ' + track.det_class, (167, 146, 11))

        cv2.putText(frame, "https://blog.csdn.net/zhaocj", (25, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        videoWriter.write(frame)

        # 如果車輛在紅燈狀態下超過停止線，保存前5秒的視頻
        if car_cross_stop_line:
            save_video(frame_buffer, "/content/drive/MyDrive/violation-video.mp4", fps, size, fourcc)
            # 清空緩存以避免重複保存相同的視頻段
            frame_buffer.clear()

cap.release()
videoWriter.release()
cv2.destroyAllWindows()
```

```python
#測試框車牌
!pip install ultralytics
!pip install deep-sort-realtime
import numpy as np
import cv2
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort
from collections import deque
from google.colab.patches import cv2_imshow

# 加载微调后的YOLOv8模型
model = YOLO('yolov8l.pt')

# 打開視頻文件
cap = cv2.VideoCapture('/content/drive/MyDrive/0807.mov')  # 輸入影片
fps = cap.get(cv2.CAP_PROP_FPS)
size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
fNUMS = cap.get(cv2.CAP_PROP_FRAME_COUNT)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
videoWriter = cv2.VideoWriter("/content/drive/MyDrive/test-video05.mp4", fourcc, fps, size) # 輸出影片

# 初始化DeepSort跟蹤器
tracker = DeepSort(max_age=30)

def box_label(image, box, label='', color=(128, 128, 128), txt_color=(0, 255, 0)):
    p1, p2 = (int(box[0]), int(box[1])), (int(box[2]), int(box[3]))
    cv2.rectangle(image, p1, p2, color, thickness=2, lineType=cv2.LINE_AA)
    if label:
        w, h = cv2.getTextSize(label, 0, fontScale=2 / 3, thickness=1)[0]
        outside = p1[1] - h >= 3
        p2 = p1[0] + w, p1[1] - h - 3 if outside else p1[1] + h + 3
        cv2.rectangle(image, p1, p2, color, -1, cv2.LINE_AA)
        cv2.putText(image,
                    label, (p1[0], p1[1] - 2 if outside else p1[1] + h + 2),
                    0, 2/3, txt_color, thickness=1, lineType=cv2.LINE_AA)
        # 在框框上方顯示座標
        coord_label = f"({int(box[0])}, {int(box[1])}) - ({int(box[2])}, {int(box[3])})"
        cv2.putText(image, coord_label, (p1[0], p1[1] - 22 if outside else p1[1] + h + 22),
                    0, 2/3, txt_color, thickness=1, lineType=cv2.LINE_AA)

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    results = model(frame, conf=0.4)
    outputs = results[0].boxes.data.cpu().numpy()

    detections = []
    red_light_detected = False
    stop_line_detected = False
    car_cross_stop_line = False

    if outputs is not None:
        for output in outputs:
            x1, y1, x2, y2 = list(map(int, output[:4]))
            label = int(output[5])
            # 假設類別2是車，類別5是公交車，類別7是卡車，類別10是紅綠燈，類別11是停止線，類別12是車牌
            if label == 2:
                detections.append(([x1, y1, int(x2-x1), int(y2-y1)], output[4], 'car'))
            elif label == 5:
                detections.append(([x1, y1, int(x2-x1), int(y2-y1)], output[4], 'bus'))
            elif label == 7:
                detections.append(([x1, y1, int(x2-x1), int(y2-y1)], output[4], 'truck'))
            elif label == 10:  # 紅綠燈
                red_light_detected = output[6] == 0  # 假設紅燈標記為0
                box_label(frame, (x1, y1, x2, y2), 'Red Light', color=(0, 0, 255), txt_color=(0, 255, 0))  # 紅色框
            elif label == 11:  # 停止線
                stop_line_detected = True
                stop_line = (x1, y1, x2, y2)
                box_label(frame, (x1, y1, x2, y2), 'Stop Line', color=(255, 255, 0), txt_color=(0, 255, 0))  # 黃色框
            elif label == 12:  # 車牌
                box_label(frame, (x1, y1, x2, y2), 'Plate', color=(0, 255, 0), txt_color=(0, 0, 0))  # 綠色框

        tracks = tracker.update_tracks(detections, frame=frame)

        for track in tracks:
            if not track.is_confirmed():
                continue
            track_id = track.track_id
            bbox = track.to_ltrb()

            # 檢查車輛是否超過停止線
            if stop_line_detected and red_light_detected:
                car_center_x = (bbox[0] + bbox[2]) // 2
                car_center_y = (bbox[1] + bbox[3]) // 2
                if car_center_y > stop_line[1]:  # 檢查車輛是否超過停止線
                    car_cross_stop_line = True

            box_label(frame, bbox, '#' + str(int(track_id)) + ' ' + track.det_class, (167, 146, 11))

        cv2.putText(frame, "https://blog.csdn.net/zhaocj", (25, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        videoWriter.write(frame)

    else:
        break

cap.release()
videoWriter.release()
cv2.destroyAllWindows()
```
